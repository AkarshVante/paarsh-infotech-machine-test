```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:\Paarsh mt\Job market and employment.csv')
```

```python
df.head()
```

```python
selected_columns = ['Age', 'Gender', 'Highest Education Level']
print("Data types of columns:\n", df[selected_columns].dtypes)
print("\nMissing values count :\n", df[selected_columns].isnull().sum())
```

```python
print("Unique values in 'Age' column:\n", df['Age'].unique())
```

```python
age_mapping = {'Thirty': '30'}
df['Age'] = df['Age'].replace(age_mapping)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
print("Data types after 'Age' conversion:\n", df['Age'].dtype)
print("Unique values in 'Age' after conversion and coercion:\n", df['Age'].unique())
```

```python
print(df.select_dtypes(include=np.number).describe())
```
